```python

import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('cars1.csv')
```

```python
data
```

```python
data.info()
```

```python
#smazani prazdnych sloupcu
df=data.drop(columns=['Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'])
```

```python
df
```

```python
#u dvou radku byla uvedena hodnota ? misto cisla u horsepower
# prevedeni ? na 0 abychom mohli prevest type object  na numeric

df['horsepower']=df['horsepower'].str.replace('?', '0')
```

```python
df.loc[32]
```

```python
df.loc[126]
```

```python
#prevod sloupce "horsepower" z typu object na numeric (int)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
df.info()
```

```python
#spocitani prumeru s vynechanim nulovych hodnot + prevedeni na integer
med_value = int(df[df['horsepower'] != 0]['horsepower'].median())

#nahrazeni '0' za prumer hodnot horsepower
df['horsepower'] = df['horsepower'].replace(0, med_value)
```

```python
#linearni regrese

X = df[['weight']]
y = df['mpg']
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# vytvořeni instance třídy LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from IPython.display import display, Latex

beta0 = round(model.intercept_, 2)
beta1 = round(model.coef_[0], 2)

display(Latex(f'$y_i = \\beta_0 + \\beta_1 \cdot x_i = {beta0} + {beta1} \cdot x_i $'))
```

```python
Xpred = pd.DataFrame({
    'weight': [2500]
})
model.predict(Xpred)  # jedno pozorování
```

```python
model.predict([[2500], [7240]])  # dvě pozorování
```

```python
X_test = np.linspace(start=1800, stop=6000).reshape(-1, 1)
y_test = model.predict(X_test)
```

```python
sns.scatterplot(
    data=df,
    x='weight',
    y='mpg'
)
plt.plot(
    X_test,
    y_test,
    c='red',
    label=f'$y_i = {beta0} + {beta1} \cdot area $'
)

plt.legend()
plt.show()
```

```python
from sklearn.metrics import  mean_squared_error,r2_score
```

```python
y_predict=model.predict(X)
MSE_1=mean_squared_error(y, y_predict)
R2_1=r2_score(y, y_predict)
print(f"MSE: {MSE_1}")
print(f"R2: {R2_1}")
```
